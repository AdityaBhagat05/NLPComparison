```python
import spacy
import pandas as pd
```

```python
df_names = pd.read_csv("Names\Dataset-Indian-Names-master\Indian_Names.csv")
names = df_names["Name"].tolist() 
print(names[:10])
```

```python
df_cities = pd.read_csv("Cities\cities_r2.csv")
cities = df_cities["name_of_city"].tolist() 
print(cities[:10])
```

```python
df_orgs = pd.read_csv(r"Orgs\registered_companies.csv", nrows=10000)
orgs = df_orgs["COMPANY_NAME"].tolist() 
print(orgs[:10])
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
tagged_sents = treebank.tagged_sents()[:10000]
```

```python
import random
```

```python
augmented_tagged_sents=[]


for sent in tagged_sents:
    words, tags = zip(*sent)
    text = " ".join(words)
    doc = nlp(text)
    new_words = list(words)

    for ent in doc.ents:
        replacement = None
        if ent.label_ == "PERSON":
            replacement = random.choice(names)
            start = ent.start
            if start < len(new_words):
                new_words[start] = replacement
        elif ent.label_ == "GPE":
            replacement = random.choice(cities)
        elif ent.label_ == "ORG":
            replacement = random.choice(orgs)
        if replacement and ent.label_ in {"GPE", "ORG"}:
            start, end = ent.start, ent.end
            repl_tokens = replacement.split()
            span_len = end - start
            for i in range(span_len):
                if start + i < len(new_words):
                    if i < len(repl_tokens):
                        new_words[start + i] = repl_tokens[i]
                    else:
                        new_words[start + i] = ""

    final_sent = [(w, t) for w, t in zip(new_words, tags) if w != ""]
    augmented_tagged_sents.append(final_sent)
```

```python
for i in range(5):
    print("Original:", tagged_sents[i])
    print("Augmented:", augmented_tagged_sents[i])
    print()
```

```python
import pickle
random.shuffle(augmented_tagged_sents)
split = int(0.8 * len(augmented_tagged_sents))
train_data = augmented_tagged_sents[:split]
test_data = augmented_tagged_sents[split:]
with open("train_pos_data.pkl", "wb") as f:
    pickle.dump(train_data, f)

with open("test_pos_data.pkl", "wb") as f:
    pickle.dump(test_data, f)

print(f"Saved {len(train_data)} training and {len(test_data)} testing sentences.")
```
